```python
import numpy as np
import matplotlib.pyplot as plt
```

$$ \frac{1}{2} \frac{1}{s^2 - 2} - \frac{1}{4} ln \left( 1 + \frac{2}{s^2 - 2} \right) + \frac{1}{4} = C'$$

```python
g = lambda x: 2/(x**2 - 2)
f_right = lambda x: g(x) - np.log(1 + g(x))
f_left  = lambda x:-g(x) + np.log(-g(x) - 1)
```

```python
delta = 0.01
x_right = np.linspace(np.sqrt(2) + delta, 5, 100)
x_left  = np.linspace(0.1, np.sqrt(2) - delta, 100)
```

```python
fig,ax = plt.subplots(1,1,dpi=200)

ax.plot(x_right, f_right(x_right),color="black")
ax.plot(x_left, f_left(x_left), color="black")

ax.plot((0,0),(-4,9),color="black",linestyle="--",lw=0.75,alpha=0.75)
ax.plot((np.sqrt(2),np.sqrt(2)),(-4,9),color="black",linestyle="--",lw=0.75,alpha=0.75)
ax.plot((-0.1,4),(3,3),color="blue",lw=1)

ax.set_ylabel("F(s)", fontsize=15)
ax.set_xlabel("s", fontsize=15)

ax.set_xlim(-0.1,4)
ax.set_ylim(-4,9)

plt.show()
```
